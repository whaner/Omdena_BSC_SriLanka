```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
# Load the keywords from a CSV file into a DataFrame
keywords_df = pd.read_csv('Data Sources And Keywords - Keywords.csv')
```

```python
keywords_df_only_Keyword=keywords_df["Keyword"]
keywords_df_only_Keyword
```

```python
# Initialize a list to store dictionaries
result_data = []

# Iterate through each keyword in the DataFrame
for keyword in keywords_df_only_Keyword:
    
    payload = {'searchword': keyword,'ordering':'newest','searchphrase':'any','Itemid':'122','option':'com_search',
                'lang':'en','limit':'0'}
    r = requests.get('https://www.aidscontrol.gov.lk/', params=payload, headers=headers)
    if r.status_code == 200:
        soup = BeautifulSoup(r.content, "html.parser")
        # Select all <a> elements within <dt> elements with the class 'result-title'
        result_title_links = soup.select('dt.result-title a')
        # Extract href attributes and anchor text, and store them in the DataFrame
        for link in result_title_links:
            href = link.get('href')
            anchor_text = link.text.strip()
            result_data.append({'Keyword': keyword, 'Anchor Text': anchor_text, 'Href': href})

    time.sleep(random.uniform(1, 2))

    # Create the DataFrame from the list of dictionaries
    result_df = pd.DataFrame(result_data)

# Print or further process the resulting DataFrame
print(result_df.head())
```

```python
duplicate_result_df=result_df
duplicate_result_df
```

```python
import csv
data = []
seen_headings = set()

for url in duplicate_result_df:
    r=requests.get(url,headers=headers)
    if r.status_code==200:
        soup = BeautifulSoup(r.content, "html.parser")
        # print(soup)
        headings = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

    # Find all paragraph elements
        paragraphs = soup.find_all('p')

        url_data = {'URL': url, 'Headings': '', 'Paragraphs': ''}


        for heading in headings:
            heading_text = heading.get_text(strip=True)
            if heading_text not in seen_headings:
                url_data['Headings'] += heading_text + ' '
                seen_headings.add(heading_text)

    # Add the paragraphs to the data
        for paragraph in paragraphs:
            url_data['Paragraphs'] += paragraph.get_text(strip=True) + ' '

    # Add the data for this URL to the list
        data.append(url_data)
```

```python
df=pd.DataFrame(data)
df.to_csv('output1.csv',index=False)
```
