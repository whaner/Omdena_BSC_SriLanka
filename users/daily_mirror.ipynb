## Loading Data

###  DAILY MIRROR


```python
# Use request package to make a GET request for the website.

req =requests.get('https://www.dailymirror.lk/')
```

```python
# using the correct encoding of html page

req.encoding = 'utf-8'
```

```python
# Extract the html
html = req.text
```

```python
# Printing the some characters in html
print(html[:50000])
```

```python
# Create a BeautifulSoup object
soup = BeautifulSoup(html)
```

```python
# Get the text out of the soup
text = soup.get_text()
```

```python
len(text)
```

```python
# See some text

print(text[:2000])
```

```python
# Show the entire text

text
```

```python
clean_text= text.replace("\n", " ")
```

```python
clean_text= clean_text.replace("/", " ")
```

```python
clean_text= ''.join([c for c in clean_text if c != "\'"])
```

```python
clean_text
```

```python
# Get the sentences

sentence=[]


tokens = nlp(clean_text)
for sent in tokens.sents:
    sentence.append((sent.text.strip()))
```

```python
sentence
```

```python
print(len(sentence))
```

```python
# Using TextBlob
textblob_sentiment=[]

for s in sentence:
    txt= TextBlob(s)

    textblob_sentiment.append([s])
```

```python
df_textblob = pd.DataFrame(textblob_sentiment, columns =['Sentence'])
```
