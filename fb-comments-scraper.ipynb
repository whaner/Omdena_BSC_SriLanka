```python
import csv
import json

# Specify the CSV file path
csv_file_path = 'content/fb_output_raw_202401240133.csv'

# Open the CSV file in read mode
with open(csv_file_path, 'r', newline='', encoding='utf-8') as csv_file:
    # Create a CSV reader object
    csv_reader = csv.reader(csv_file)

    # Initialize a list to store the dictionaries
    data_list = []

    # Iterate through each row in the CSV file and make dictionary of list
    for row in csv_reader:
      
        json_string = row[0]

       
        record = json.loads(json_string)

     
        data_list.append(record)


for record in data_list:
    print(record)
```

```python
posts_with_comments = []
for record in data_list:
    comments_count = record.get('comments', '')
    if comments_count > 0 :
       posts_with_comments.append(record)
        
```

```python
clean_posts_with_comments = []
    # Iterate through the list of dictionaries and write each row
for record in posts_with_comments:
    row = {
        'timestamp': record.get('timestamp',''),
        'postID': record.get('postId',''),
        'url': record.get('url', ''),
        'pageName': record.get('pageName', ''),
        'Text': record.get('text', ''),
        'ocrText': record.get("media", [{}])[0].get("ocrText", ''),
        'likes': record.get('likes', ''),
        'comments': record.get('comments', ''),
        'shares': record.get('shares', ''),
        'Media_Type': record.get("media", [{}])[0].get("__typename", '')
            }
    clean_posts_with_comments.append(row)
```

```python
import pandas as pd
# Convert the list of dictionaries to a DataFrame
df_clean_posts_with_comments = pd.DataFrame(clean_posts_with_comments)

# Print the DataFrame
print(df_clean_posts_with_comments[['postID','url','comments']].head(5))
```

```python
# Load API key from JSON file
with open('content/apify_key.json', 'r') as file:
    api_key_data = json.load(file)
    api_key = api_key_data.get('key_omdena')  # Replace 'api_key' with the actual key field in your JSON
```

```python
from apify_client import ApifyClient

# Initialize the ApifyClient with your API token
client = ApifyClient(api_key)

# Prepare the Actor input
run_input = {
    "startUrls": dict_selected_posts,
    "resultsLimit": 10,
    "includeNestedComments": False,
    "viewOption": "RANKED_UNFILTERED",
}

# Run the Actor and wait for it to finish
run = client.actor("us5srxAYnsrkgUv2v").call(run_input=run_input)

# Fetch and print Actor results from the run's dataset (if there are any)
#for item in client.dataset(run["defaultDatasetId"]).iterate_items():
#   print(item)
```

```python
fb_comments =[]
# Fetch and print Actor results from the run's dataset (if there are any)
for record in client.dataset(run["defaultDatasetId"]).iterate_items():
    row = {
        'postURL': record.get('facebookUrl',''),
        'datetime': record.get('date',''),
        'Text': record.get('text','')
            }
    fb_comments.append(row)
```
