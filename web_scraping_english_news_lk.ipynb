```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

The User-Agent string provided is crafted to mimic a common user agent for the Google Chrome browser running on a Windows 10 operating system.

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
# Load the keywords from a CSV file into a DataFrame
keywords_df = pd.read_csv('english_keywords.csv')

# Initialize a list to store dictionaries
result_data = []

# Iterate through each keyword in the DataFrame
for index, row in keywords_df.iterrows():
    keyword = row[1]  # Assuming 'keyword' is the column name in your CSV file
    payload = {'searchword': keyword,'limit':'0'}
    r = requests.get('https://news.lk/component/search', params=payload, headers=headers)
    if r.status_code == 200:
        soup = BeautifulSoup(r.content, "html.parser")
        # Select all <a> elements within <dt> elements with the class 'result-title'
        result_title_links = soup.select('dt.result-title a')
        # Extract href attributes and anchor text, and store them in the DataFrame
        for link in result_title_links:
            href = link.get('href')
            anchor_text = link.text.strip()
            result_data.append({'Keyword': keyword, 'Anchor Text': anchor_text, 'Href': href})

    time.sleep(random.uniform(1, 2))

# Create the DataFrame from the list of dictionaries
result_df = pd.DataFrame(result_data)

# Print or further process the resulting DataFrame
print(result_df.head())
```

```python
# Save the DataFrame to a CSV file
result_df.to_csv('/content/result_data.csv', index=False,encoding='utf-8')
```

```python
import json
article_data = []
for index,row in urls.iterrows():
    url = row['url']
    r = requests.get(url, headers=headers)
    if r.status_code == 200:
        soup = BeautifulSoup(r.content, "html.parser")
        script_tag = soup.find('script', type='application/ld+json')

            # Check if the script tag is found
        if script_tag:
          # Convert the script content to a dictionary
          script_content = json.loads(script_tag.string)
          if 'articleBody' in script_content:
            # Print the articleBody
            article_body = script_content['articleBody']
            print(article_body)
            article_data.append({'URL': url, 'Title': soup.title.string, 'Text': article_body })
          else:
            print("articleBody not found in the script content.")
        else:
            print("Script tag not found.")
    else:
        print(f"Failed to retrieve content. Status Code: {random.status_code}")
        time.sleep(random.uniform(1, 2))

# Create the DataFrame from the list of dictionaries
result_df = pd.DataFrame(article_data)

# Print or further process the resulting DataFrame
print(result_df.head())
```

```python
# Save the DataFrame to a CSV file
result_df.to_csv('/content/newslk_sinhala.csv', index=False,encoding='utf-8')
```
